What are those `\r\n` things? Well, recall we talked about line feeds and carriage returns. That is what we are seeing. Ack. And then we aren't taking into consideration punctuation. So two problems:

1. How can we **remove** newline things?
2. How can we tokenize punctuation?

For the first, notice that we always get the return `\r` then the newline `\n`. So we can just replace that. With what? If we replace it with nothing, then words will run together. No, it has to be another space:

    nonewlines = text.replace("\r\n", " ")

For the punctuation, we want to replace, say the ")" with " ) ". Then it would split ok. Let's do that.

    noparens = nonewlines.replace(")", " ) ")

This will work, but we'll need to do it for every piece of punctuation. What we'd like is a way to do the above for **each** type of punctuation. Something like:

1. Make a list of all cases of punctuation.
2. For each punctuation mark in the list:
    a. form the replacement string " " + mark + " "
    b. replace the mark with the replacement.
    c. repeat.

The way we go through the list, replacing over and over gives us our first [control flow](http://en.wikipedia.org/wiki/Control_flow) construct, the **for loop**. We will talk about two cases today:

```python
import urllib2

def main():
    url = "http://www.gutenberg.org/cache/epub/582/pg582.txt"
    returned = urllib2.urlopen(url)    # store contents of webpage                                 
    text = returned.read()             # extract the text from the webpage
    normedText = text.replace("\r\n", " ") # get rid of new lines
    
    # now we make sure that punctuation marks are their own tokens
    punctuation = ['!', ';', ',', '.', "'", '"', "_", ")", "(", "[", "]"]

    for punct in punctuation:
        #flank each piece of punctuation with spaces
        newPunct = " " + punct + " " 
        normedText = normedText.replace(punct, newPunct)

    print normedText[0:150]
    # we might get multiple blanks; get rid of those
    # this could be a while loop...
    normedText = normedText.replace("  ", " ")
    
    tokens = normedText.split(" ")
    print tokens[0:50]

    
if __name__ == "__main__":
    main()
```

If we want to get anything reasonable out of this, like an ordered list of tokens, it will be a little complicated. We can definitely do it. But for now, let's not reinvent the wheel.

The Python Standard Library includes the `Counter` data type (class), which is a version of a dict that can show us counts. It also handles the if then stuff quite well. Let's use it.

```python
import urllib2
from collections import Counter

def main():
    url = "http://www.gutenberg.org/cache/epub/582/pg582.txt"
    returned = urllib2.urlopen(url)    # store contents of webpage                                 
    text = returned.read()             # extract the text from the webpage
    normedText = text.replace("\r\n", " ") # get rid of new lines
    
    # now we make sure that punctuation marks are their own tokens
    punctuation = ['!', ';', ',', '.', "'", '"', "_", ")", "(", "[", "]"]

    for punct in punctuation:
        #flank each piece of punctuation with spaces
        newPunct = " " + punct + " " 
        normedText = normedText.replace(punct, newPunct)

    # we might get multiple blanks; get rid of those
    # this could be a while loop...
    normedText = normedText.replace("  ", " ")
    
    tokens = normedText.split(" ") #now we have our tokens
    frequencies = Counter() # this is now an instance of the Counter class
    
    # loop through and add them
    for token in tokens:
        frequencies[token] += 1
    
    print frequencies.most_common(30)
    
if __name__ == "__main__":
    main()
```

```python
import urllib2
from collections import Counter

def main():
    url = "http://www.gutenberg.org/cache/epub/582/pg582.txt"
    returned = urllib2.urlopen(url)    # store contents of webpage                                 
    text = returned.read()             # extract the text from the webpage
    normedText = text.replace("\r\n", " ") # get rid of new lines
    
    # now we make sure that punctuation marks are their own tokens
    punctuation = ['!', ';', ',', '.', "'", '"', "_", ")", "(", "[", "]"]

    for punct in punctuation:
        #flank each piece of punctuation with spaces
        newPunct = " " + punct + " " 
        normedText = normedText.replace(punct, newPunct)

    # we might get multiple blanks; get rid of those
    # this could be a while loop...
    normedText = normedText.replace("  ", " ")
    
    tokens = normedText.split(" ") #now we have our tokens
    frequencies = Counter() # this is now an instance of the Counter class
    
    # loop through and add them
    for token in tokens:
        if token != '':
            normedToken = token.lower()
            frequencies[normedToken] += 1
    
    print frequencies.most_common(30)
    
if __name__ == "__main__":
    main()
```

Great. We are almost done. Let's now save this dictionary somewhere. When we save things, we have to think about the encoding. In this case, we will use JSON, which is compact and used all over the place. To do that, we will use Python's json module

```python
import urllib2
import json
from collections import Counter

def main():
    url = "http://www.gutenberg.org/cache/epub/582/pg582.txt"
    returned = urllib2.urlopen(url)    # store contents of webpage                                 
    text = returned.read()             # extract the text from the webpage
    normedText = text.replace("\r\n", " ") # get rid of new lines
    
    # now we make sure that punctuation marks are their own tokens
    punctuation = ['!', ';', ',', '.', "'", '"', "_", ")", "(", "[", "]"]

    for punct in punctuation:
        #flank each piece of punctuation with spaces
        newPunct = " " + punct + " " 
        normedText = normedText.replace(punct, newPunct)

    # we might get multiple blanks; get rid of those
    # this could be a while loop...
    normedText = normedText.replace("  ", " ")
    
    tokens = normedText.split(" ") #now we have our tokens
    frequencies = Counter() # this is now an instance of the Counter class
    
    # loop through and add them
    for token in tokens:
        if token != '':
            normedToken = token.lower()
            frequencies[normedToken] += 1
    
    print frequencies.most_common(30)
    f = open("frequencies.json", "w")
    json.dump(frequencies, f)
    
if __name__ == "__main__":
    main()
```
