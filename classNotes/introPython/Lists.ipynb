```python
#custom sorting: use a function to get key values
names = ["Marina", "von Abel", "Antony"]
print names
names.sort()
print names #fun fact: von names should be sorted without the von...
```

Key functions take in an object and return some key of that object    

```python
def removeVon(word):
    if "von " in word:
        return word[4:]
    else:
        return word

names.sort(key=removeVon)
print names
```

Looping
-------
A key aspect of using a list. `for` loops are your best bet.

```python
li = [1,2,"4", "10"]
#you can loop over the elements themselves
for el in li:
    print int(el)+1.5

#or over the indices themselves
for index in [0,1,2,3]:
    print int(li[index])+2.5

for i in range(0,len(li)):
    print int(li[i])+3.5
```

# Functional Programming Tricks

When looping, we are often trying to get a new version of a list. We do this by appending


```python
tokens = "A tree that Mary told me about is in Boston".split(" ")
uncasedTokens = []
for token in tokens:
    uncasedTokens.append(token.lower())
print sorted(uncasedTokens)
```

But Python allows **list comprehensions**, analogs of set builder notation to do these procedures in one line.

```python
uncasedTokens = [token.lower() for token in tokens]
print uncasedTokens

mixedTokenPairs = [(tokens[i].lower(), tokens[i+1].upper()) for i in range(0,len(tokens)-1)]
print mixedTokenPairs
```

```python
# multiple list comprehensions, read from left to right
letters = [letter for token in tokens for letter in token]

# this is:
# ret = []
# for token in tokens:
#     for letter in token:
#         ret.append(letter)

print letters
```

* Python also allows a limited set of operations that apply functions to elements of a list sequentially and return the results
    - `map`: applies a function to each element, returns the list of the return values
    - `filter`: applies a boolean function to each element, ignores those that return False


```python
import string

uncased2 = map(string.lower, tokens)
print uncased2

def isntFunctionWord(word):
#    if word in [....]:
#        return False
#    else:
#        return True
    return word not in ["a", "that", "the", "a", "is", "was", "were"]

filtered = filter(isntFunctionWord, uncased2)
print filtered
```
