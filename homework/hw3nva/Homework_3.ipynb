## Problem 1

Your job in the first problem is simply to load the file, and tokenize it. The relevant files to choose are in the data/SampleTexts directory. There are two ways of saying where that is. So, ".." means the parent of HW3, which is homework, and "../.." is the github repo root.

Use relative paths to open one of the files in SampleTexts. Write another function that reads through the text, line by line, and tokenizes it. Since we are going for phonemes, you can ignore any tokens not in the CMU dictionary, including punctuation. Your tokenizer should examine the CMU dictionary and return a list of tokens, where each token is the CMU pronunciation for that word. If the word has multiple pronunciations, choose the first. If the word isn't in the dictionary, just ignore it.

```python
import json

def tokenizeFile(filename):
    ''' takes a path to a file, and returns 
    a list of words in the file'''
    # Load the file
    f = open(filename, "r")
    lines = " ".join(f.readlines())
    f.close()
    del f 
    
    # Make everything lowercase
    lines = lines.lower()
    
    # Remove Punctuation, numbers, and \r
    punctuation = "`~,.?/\\'\";:!#$%^&*()_+=[]{}|\r0123456789" 
    for symb in punctuation:
        lines = lines.replace(symb, "")
    
    # Replace \n- with nothing
    lines = lines.replace("\n-", "")
    
    # Replace \n and - with space
    lines = lines.replace("\n", " ")
    
    # We make a chioce to treat "wood-chuck" as "wood chuck"
    lines = lines.replace("-", " ")
    
    # Split on Empty space
    wordList = lines.split()
        
    return wordList

def trimWordList(wordList):
    ''' trims words from a wordlist that are not in the CMU dict
    '''
    
    # load the dictionary
    dictFile = open("../Data/cmdict7a.json", "r")
    cmuDict = json.load(dictFile)
    dictFile.close()
    del dictFile
    
    # Get rid of every word that isn't in the dictionary
    finalWordList = []
    for word in wordList:
        if word in cmuDict:
            # Add the first pronunciation of every word to the wordlist
            entry = cmuDict[word][0]
            finalWordList.append(entry)
    return finalWordList

if __name__ == "__main__":
    wordList = tokenizeFile("../Data/grimms")
    wordList = trimWordList(wordList)
    print wordList[:101]



```

## Problem 2

Now, write another function that loops through the tokens and, for each word in turn, counts the number of times pairs of phonemes appear. Specifically, if a line was "how much wood could a wood chuck chuck if a woodchuck could chuck wood"

with pronunciations


[u'HH', u'AW'] [u'M', u'AH', u'CH'] [u'W', u'UH', u'D'] [u'K', u'UH', u'D'] [u'AH'] [u'W', u'UH', u'D'] [u'CH', u'AH', u'K'] [u'CH', u'AH', u'K'] [u'IH', u'F'] [u'AH'] [u'W', u'UH', u'D', u'CH', u'AH', u'K'] [u'K', u'UH', u'D'] [u'CH', u'AH', u'K'] [u'W', u'UH', u'D']

we want a dict of dicts that ends up like this:

{u'AH': {u'CH': 1, u'K': 4},
 u'CH': {u'AH': 4},
 u'D': {u'CH': 1},
 u'HH': {u'AW': 1},
 u'IH': {u'F': 1},
 u'K': {u'UH': 2},
 u'M': {u'AH': 1},
 u'UH': {u'D': 6},
 u'W': {u'UH': 4}}

```python
import json

def tokenize(sentence):
    ''' takes a sentence and returns a list of words in that sentence
    '''
    # Split the sentence
    wordList = sentence.split(" ")
    lines = " ".join(wordList)
    
    # Make everything lowercase
    lines = lines.lower()
    
    # Remove Punctuation, numbers, and \r
    punctuation = "`~,.?/\\'\";:!#$%^&*()_+=[]{}|\r0123456789" 
    for symb in punctuation:
        lines = lines.replace(symb, "")
    
    # Replace \n- with nothing
    lines = lines.replace("\n-", "")
    
    # Replace \n and - with space
    lines = lines.replace("\n", " ")
    
    # We make a chioce to treat "wood-chuck" as "wood chuck"
    lines = lines.replace("-", " ")
    
    # Split on Empty space
    wordList = lines.split()
        
    return wordList

def trimSentence(sentence):
    ''' trims words from a sentence that are not in the CMU dict after tokenizing the sentence
    '''
    # load the dictionary
    dictFile = open("../Data/cmdict7a.json", "r")
    cmuDict = json.load(dictFile)
    dictFile.close()
    del dictFile
    
    # Get rid of every word that isn't in the dictionary
    finalWordList = []
    wordList = tokenize(sentence)
    for word in wordList:
        if word in cmuDict:
            # Add the first pronunciation of every word to the wordlist
            entry = cmuDict[word][0]
            finalWordList.append(entry)
    return finalWordList

def bigramize(wordList):
    '''defines a bigram dictionary based tokenized input
    '''
    # Take the sentence and chop it up into the dictionary items for each word
    bigram = {}
    # Go through every word in the sentence and update the bigram dictionary
    for word in wordList:
        for i in range(0,(len(word)-1)):
            phoneme1 = word[i]
            phoneme2 = word[i+1]
            # Case 1: phoneme1 is in the bigram dictionary already
            if phoneme1 in bigram:
                # Subcase 1A: the following phoneme is already a bigram dictionary entry, update by 1
                if phoneme2 in bigram[phoneme1]:
                    bigram[phoneme1][phoneme2] += 1
                # Subcase 1B: the following phoneme isn't yet in the bigram dict, define it as 1
                else:
                    bigram[phoneme1][phoneme2] = 1
                # Case 2: phoneme1 is not yet in the dictionary, define phoneme2 relative to it
            else:
                bigram[phoneme1] = {phoneme2 : 1}
    return bigram	

testtoken = trimSentence("how much wood could a wood chuck chuck if a woodchuck could chuck wood")
print bigramize(testtoken) == {u'AH': {u'CH': 1, u'K': 4}, u'CH': {u'AH': 4}, u'D': {u'CH': 1}, u'HH': {u'AW': 1}, u'IH': {u'F': 1}, u'K': {u'UH': 2}, u'M': {u'AH': 1}, u'UH': {u'D': 6}, u'W': {u'UH': 4}}
```

## Problem 5

Extend your class to have the following special methods:

1. a `__len__` method, which returns the number of phonemes encountered across the file,
2. a `__cmp__` method, which compares two `FrequencyTable` instances by the number of phonemes that occur in each instance, and
3. a `__getitem__` method, which returns the subdict for a given first phoneme.

For each of the 5 files in the SampleTexts directory, instantiate a class, and show the lengths and contingency tables from each except. Put them in a list and sort them in reverse.

```python
class FrequencyTable(object):
    def __init__(self, filename):
        # use existing functions to tokenize, trim, and generate the frequency table
        tokenizedFile = tokenizeFile(filename)
        tokenizedFile = trimWordList(tokenizedFile)
        self.frequencies = bigramize(tokenizedFile)
        
        # so we can check to make sure that sorting works
        self.filename = filename
    def __str__(self):
        dictionary = self.frequencies
        
        def makeHeader():
            ''' gets a combined list of 
            '''
            # Grab the list of phoneme1's
            header = dictionary.keys()
        
            # Get all the second phoneme's
            # For every first phoneme in the dictionary
            for phoneme1 in dictionary:
                # ....define a subdictionary of those phonemes within phoneme1's dictionary
                subDict = dictionary[phoneme1]
                for phoneme2 in subDict:
                    # Go through every phoneme2 and if it's not already in the header list, append it
                    if phoneme2 not in header:
                        header.append(phoneme2)
                        
            # Alphabetize the header list
            header.sort()
            return header
    
        def makeRow(phoneme2):
            ''' makes a row
                !Assumes header is already defined!
            '''
            # initilize the row as the empty string
            row = "\n"
        
            # Get the stuff in the row
            for phoneme1 in header:
                # Check every phone in the list against one another
                if phoneme1 in dictionary:
                   if phoneme2 in dictionary[phoneme1]:
                        # Add string of the bigram freq for every phoneme1 that has a frequency for phoneme2
                        row = row + '%6s' % dictionary[phoneme1][phoneme2]
                    # Add empty character to the string if phoneme1 has no entry for phoneme2
                   else: 
                        row = row + '%6s' % " "
                # Add empty character for every element in header that has no entry for being phoneme1
                else:
                   row = row + '%6s' % " "
               
            # Create row labels
            row = row + '%4s' % phoneme2
            return row
        
        header = makeHeader()
    
        # Print the top line
        topline = ""
        for phoneme in header:
            topline = topline + '%6s' % phoneme
        
        #Print the rows
        string = topline
        for phoneme in header:
            string = string + makeRow(phoneme)
        return string

    def __len__(self):
        ''' returns the number of phonemes in the FrequencyTable
        '''
        Phons = 0
        # Sum every phoneme found
        for key1 in self.frequencies:
            for key2 in self.frequencies[key1]:
                Phons = Phons + self.frequencies[key1][key2]
        # Because we sum frequencies for each phoneme1 and phoneme2, we divide by 2
        Phons = Phons / 2
        return Phons
    
    def __cmp__(self, other):
        ''' compares two FrequencyTables 
        by the number of phonemes in each instance '''
        # Instantiate frequency tables
        OwnPhonLen = self.__len__()
        OtherPhonLen = other.__len__()
        
        # Compare the lengths of two frequency tables
        # If self is shorter than what it's being compared to, return -1
        if OwnPhonLen < OtherPhonLen:
            return -1
        # If self is longer than what it's being compared to, return 1
        if OtherPhonLen > OwnPhonLen:
            return 1
        # Otherwise, self is equal to the other, return 0
        else:
            return 0
        
    def __getitem__(self, phoneme):
        ''' returns the subdict for a given first phoneme
        '''
        return self.frequencies[phoneme]
    
# Create frequency tables for all of the texts
grimms = FrequencyTable("../Data/grimms")
mountainhouses = FrequencyTable("../Data/mountainhouses.txt")
twain = FrequencyTable("../Data/twain")
yellowwallpaper = FrequencyTable("../Data/yellowwallpaper")
analyze_people = FrequencyTable("../Data/analyze_people")

# Create a list of all frequency tables in order to combine them
mylist = [grimms, mountainhouses, twain, yellowwallpaper, analyze_people]

# Go through every frequency table in the list and summing all of its phonemes
for i in mylist:
    # For each text in list, print the filename and the sum of its bigrams
    print i.filename, len(i)

# Sort each frequency table in reverse
mylist.sort(reverse = True)

# Print the reverse list of frequency tables for all of the texts.
for i in mylist:
    print i.filename, len(i)


print grimms
print mountainhouses
print twain
print yellowwallpaper
print analyze_people


```
